## Data Wrangling

### Gathering Data

#### data kualitas udara seluruh stasiun

```python
Aotizhongxin = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv")
Changping = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Changping_20130301-20170228.csv")
Dingling = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Dingling_20130301-20170228.csv")
Dongsi = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Dongsi_20130301-20170228.csv")
Guanyuan = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Guanyuan_20130301-20170228.csv")
Gucheng = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Gucheng_20130301-20170228.csv")
Huairou = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Huairou_20130301-20170228.csv")
Nongzhanguan = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Nongzhanguan_20130301-20170228.csv")
Shunyi = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Shunyi_20130301-20170228.csv")
Tiantan = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Tiantan_20130301-20170228.csv")
Wanliu = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Wanliu_20130301-20170228.csv")
Wanshouxigong = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Wanshouxigong_20130301-20170228.csv")
```

```python
gabungan = pd.concat([Aotizhongxin, Changping, Dingling, Dongsi, Guanyuan, Gucheng, Huairou, Nongzhanguan, Shunyi, Tiantan, Wanliu, Wanshouxigong], axis=0)
gabungan
```

Menghapus kolom No

```python
gabungan.drop('No',axis=1, inplace=True)
```

menambahkan kolom time_stamp

```python
time_stamp = gabungan.apply(lambda row: datetime.datetime(row['year'], row['month'], row['day'], row['hour']), axis=1)
```

```python
gabungan.insert(0,'time_stamp',time_stamp)
```

```python
gabungan.head()
```

Mengubah kolom month

```python
gabungan['month']= gabungan['month'].map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})
```

```python
gabungan.head()
```

mengganti kolom tanggal menjadi hari

```python
def number_to_day(number):
  day_number = (number  - 1) % 7
  return calendar.day_name[day_number]
```

```python
gabungan['day'] = gabungan['day'].apply(number_to_day)
```
